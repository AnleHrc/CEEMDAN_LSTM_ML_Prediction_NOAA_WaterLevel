```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
def Plot_Violinplot(file):
    df = pd.read_csv('../Data/NOAA/All Station/{}'.format(file))
    df = df[['Wind Speed (m/s)','Wind Gust (m/s)', 'Air Temp (°C)', 'Water Level(m)']]

    fig, ax = plt.subplots(figsize=(12,11))
    ax.grid(False)

    colors = ['red', 'blue', 'green', 'purple']
    parts = ax.violinplot(df.values, showmeans=False, showmedians=False,
                          showextrema=False)

    for i, pc in enumerate(parts['bodies']):
        pc.set_facecolor(colors[i])
        pc.set_edgecolor('black')
        pc.set_alpha(1)

    ax.legend(parts['bodies'], df.columns,prop={'weight': 'bold', 'size': 10},title_fontsize=12, )
    ax.spines['left'].set_color('black')
    ax.spines['bottom'].set_color('black')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    ax.set_xticks([1, 2, 3, 4])
    ax.set_xticklabels(['Wind Speed (m/s)', 'Wind Gust (m/s)', 'Air Temp (°C)','Water Level(m)'],fontsize=13)

    ax.set_xlabel('Index',fontweight='bold', fontsize=15)
    ax.set_ylabel('Value',fontweight='bold', fontsize=15)
    ax.tick_params(axis='y', labelsize=13)  # 设置y轴刻度标签（数值）的字体大小
    station_name = file.split('.')[0]
    if station_name=='KAHUKUI Station':
        station_name = 'Kahului Station'
    if station_name=='LaJolla Station':
        station_name = 'La Jolla Station'
    plt.title(station_name)

    plt.savefig('{}'.format(file)+'_4.png', transparent=True,bbox_inches='tight',dpi=600)
    plt.show()
```
